```python
# plot the density for each columns
n_row = 10
n_col = 3
f,ax = plt.subplots(n_row, n_col,figsize=(20,50))
for i in range (2,train_set.shape[1]):
    h = train_set.iloc[:,i]
    ind = i- 2
    a = ax[ind//n_col][ind%n_col]
    a.hist(h,bins = 50 ,density=True)
    a.set_xlabel('values')
    a.set_ylabel('density')
    a.set_title('column_id = {k}'.format( k=i))
```
